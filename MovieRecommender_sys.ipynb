```python
# Content-based recommender system
import numpy as np
import pandas as pd
```

```python
def load_data():
    movies = pd.read_csv('tmdb_5000_movies.csv')
    credits = pd.read_csv('tmdb_5000_credits.csv')
    return movies, credits
```

```python
movies.head()
```

```python
credits.head()
```

```python
movies, credits = load_data()
```

```python
print(type(movies))
print(type(credits))
```

```python
movies= movies.merge(credits, on='title')  # This should work now
movies.head(1)# Display the first few rows of the merged DataFrame
```

```python
movies.info()
```

```python
#overview
#id
#genre
#cast
#crew
#keywords
#movie_id
movies=movies[['movie_id', 'title', 'genres', 'keywords','overview','cast','crew']]
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
#helper funs
import ast

def convert(obj):
    L = []  # Initialize an empty list
    try:
        # Use ast.literal_eval to safely evaluate the string as a Python literal
        for i in ast.literal_eval(obj):
            L.append(i['name'])  # Append the 'name' from each dictionary
    except (ValueError, SyntaxError):  # Handle potential errors from literal_eval
        print(f"Error evaluating: {obj}")
    return L  # Return the complete list
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python

def convert3(obj):
    L = []  # Initialize an empty list
    counter = 0
    try:
        # Use ast.literal_eval to safely evaluate the string as a Python literal
        for i in ast.literal_eval(obj):
            if counter!=3:
                L.append(i['name'])  # Append the 'name' from each dictionary
            else:
                break
    except (ValueError, SyntaxError):  # Handle potential errors from literal_eval
        print(f"Error evaluating: {obj}")
    return L  # Return the complete list
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
#coverting overview from string to list
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
#removing space btwn the words , problem:when we search an name two entites will show of same firstname 
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","")for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","")for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview']+movies['genres']+movies['keywords']+movies['crew']+movies['cast']
```

```python
movies.head()
```

```python
new_df = movies[['movie_id','title','tags']]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
#text vectorisation
#text -----> vector(method:Bag of words)
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Vectorize the text data
cv = CountVectorizer(max_features=5000, stop_words='english')
vectors = cv.fit_transform(new_df['tags']).toarray()

# Compute cosine similarity on the vectorized data
similarity= cosine_similarity(vectors)
```

```python
feature_names = cv.get_feature_names_out()
print(feature_names)
```

```python
sorted(list(enumerate(similarity[0])),reverse = True,key=lambda x:x[1])[1:6]
```

```python
def recommand(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse = True,key=lambda x:x[1])[1:6]
    for i in movies_list:
        print(new_df.iloc[i[0]].title)


```

```python
recommand('Batman Begins')
```
